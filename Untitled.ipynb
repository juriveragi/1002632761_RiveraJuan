```python
#%%Ejercicio5

import keyword

texto=input("texto=")

listapalabrasreservadas=keyword.kwlist
i=0

palabrasreservadas=0
while i<len(listapalabrasreservadas):
    if listapalabrasreservadas[i] in texto:
        palabrasreservadas=palabrasreservadas+1
    i+=1

print("palabras reservadas=",palabrasreservadas)
    
letrascontilde=0
for n in texto:
    codigoletrascontildes=[193,205,243,211,250,218,225,233,201,237]
    o=ord(n)
    if o in codigoletrascontildes:
        letrascontilde=letrascontilde+1
        
print("tildes=",letrascontilde)

espacios=0
for n in texto:
    if n.isspace():
        espacios=espacios+1

print("espacios=",espacios)

numeros=0
for n in texto:
    if n.isdigit():
        numeros=numeros+1

print("numeros=",numeros)

mayusculas=0
for n in texto:
    if n.isupper():
        mayusculas=mayusculas+1
        
print("mayusculas=",mayusculas)
```

```python
#%%Ejercicio7
    
longitud=int(input("longitud de la lista="))
i=0
lista=[]
while i<longitud:
   lista.append(int(input("=")))
   i=i+1

listaordenada=lista
listaordenada.sort()

if lista==listaordenada:
    print("si esta ordenada")
    i=0
    numero=int(input("numero a insertar="))
    while i<longitud:
        if numero<lista[0]:
            lista.insert(0,numero)
        elif numero>lista[len(lista)-1]:
            lista.insert(len(lista),numero)
        elif numero>lista[i-1] and numero<lista[i]:
            lista.insert(i,numero)
        i=i+1
    
    print(lista)
else: 
    print("la lista no esta ordenada")
```

```python
#%%Ejercicio8

longitud=int(input("longitud de la lista="))
numeros=[]
i=0
n=0
m=0
eliminados=0
numeromayor=0
segundonumeromayor=0

while i<longitud:
    numeros.append(int(input("=")))
    if numeros[i]>numeromayor:
        numeromayor=numeros[i]
    i=i+1

while n<longitud-eliminados:
    if numeros[n-eliminados]==numeromayor:
        numeros.pop(n-eliminados)
        eliminados=eliminados+1
    n=n+1
        
while m<longitud-eliminados:
    if numeros[m]>segundonumeromayor:
        segundonumeromayor=numeros[m]
    m=m+1
    
print(segundonumeromayor)
```

```python
#%%Ejercicio 11

altura=int(input("altura="))
anchura=int(input("anchura="))

caracter=input("caracter=")

def rectangulo(altura,anchura,caracter):
    i=0
    
    while i < altura:
        n=0
        while n < anchura:
            print(caracter,end="")
            n=n+1
        print("")
        i=i+1

rectangulo(altura,anchura,caracter)
```

```python
#%%Ejercicio 12

altura=int(input("altura="))-1


def rectangulo(altura):
    i=0
    
    while i < altura:
        n=0
        while n <= i:
            print("*",end="")
            n=n+1
        print("")
        i=i+1
        
    while i >= 0:
        n=0
        while n <= i:
            print("*",end="")
            n=n+1
        print("")
        i=i-1

rectangulo(altura)
```

```python
import random


def sumaentredosnumeros():
    aciertos=0
    while aciertos<5:
        numero1=random.randint(1,100)
        numero2=random.randint(1,100)
        print(numero1,"+",numero2,)
        respuesta=int(input("="))
        
        if respuesta==numero1+numero2:
            aciertos+=1
            print("acertaste")
        else:
            print("vuelve a intentarlo")
    print("fin del juego")
        
sumaentredosnumeros()
```

```python
#%%Ejercicio15
import random

longitud=int(input("longitud del numero, entre 2 y 9="))

def adivinaelnumero(longitud):
    numero=[]
    i=0
    
    while i<longitud:
        digito=random.randint(0,9)
        numero.append(digito)
        i=i+1
        
    print(numero)
    
    while True:
        
        aciertos=0
        numero2=input("Adivine el numero=")
        m=0
        for i in numero2:
            n=int(i)
            if n==numero[m]:
                aciertos=aciertos+1
            m=m+1
        
        if aciertos==longitud:
            print("fin del juego")
            break
        
adivinaelnumero(longitud)
```
